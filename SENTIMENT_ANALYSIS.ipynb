```python
# Specify the text column
text_column = 'text'

# Check missing values
print(df['Comment'].isnull().sum())

# Remove rows with missing text
df = df.dropna(subset=['Comment'])

print("Updated Shape:", df.shape)
```

###6. Text Cleaning
Remove:

Lowercase conversion

URLs

HTML tags

Punctuation

Numbers

Extra whitespaces

```python
def clean_text(text):
    text = str(text).lower()                     # Convert to lowercase
    text = re.sub(r'http\S+|www\S+', '', text)  # Remove URLs
    text = re.sub(r'<.*?>', '', text)           # Remove HTML tags
    text = re.sub(r'\d+', '', text)             # Remove numbers
    text = text.translate(str.maketrans('', '', string.punctuation))
    text = re.sub(r'\s+', ' ', text).strip()    # Remove extra spaces

    return text
```

Apply cleaning:

```python
df['cleaned_text'] = df["Comment"].apply(clean_text)

df[['cleaned_text']].head()
```

###7. Tokenization

Tokenization splits text into individual words.

```python
import nltk

nltk.download('punkt')
nltk.download('punkt_tab')
nltk.download('stopwords')
```

```python
def tokenize_text(text):
    return word_tokenize(text)

df['tokens'] = df['cleaned_text'].apply(tokenize_text)

df[['tokens']].head()
```

###8. Stopword Removal

Stopwords are commonly used words that usually carry little semantic meaning.

Example:

the

is

are

and

of


```python
stop_words = set(stopwords.words('english'))

def remove_stopwords(tokens):
    return [word for word in tokens if word not in stop_words]

df['processed_tokens'] = df['tokens'].apply(remove_stopwords)

df[['processed_tokens']].head()
```

###9. Join Tokens Back into Text (Optional)

Some machine learning models require processed text strings.

```python
df['processed_text'] = df['processed_tokens'].apply(lambda x: ' '.join(x))

df[['processed_text']].head()
```

###2. Feature Extraction

### 2. Feature Extraction using TF-IDF Vectorizer

In this task, we convert the preprocessed text into numerical representations that machine learning algorithms can understand. Here, we use TfidfVectorizer from Scikit-learn.

### 12.1 Import the Required Library

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

### 2.2 Apply TF-IDF Vectorization

The processed_text column generated during preprocessing will be used as input.

```python
# Initialize the TF-IDF Vectorizer
tfidf = TfidfVectorizer(
    max_features=5000,   # Limit vocabulary size
    ngram_range=(1, 1)   # Use unigrams
)

# Transform text into TF-IDF features
X_tfidf = tfidf.fit_transform(df['processed_text'])

print("TF-IDF Matrix Shape:", X_tfidf.shape)
```

## 2.3 View the Extracted Features

```python
# Display feature names
feature_names = tfidf.get_feature_names_out()

print("First 20 Features:")
print(feature_names[:20])
```

### 2.4 Convert the Sparse Matrix to a DataFrame (Optional)

This step helps visualize the numerical representation.

```python
# Convert sparse matrix to DataFrame
tfidf_df = pd.DataFrame(
    X_tfidf.toarray(),
    columns=feature_names
)

# Display first few rows
tfidf_df.head()
```

###2.5 Example of TF-IDF Transformation

```python
sample_text = df['processed_text'].iloc[0]

print("Processed Text:")
print(sample_text)

print("\nTF-IDF Vector:")
print(X_tfidf[0].toarray())
```

### 3. Model Development
a)Naive Bayes      
b)Support Vector Machine

###3.1 Define Features and Target Variable

```python
# Define target variable
target_column = 'Emotion'   # Emotion labels column

y = df[target_column]

print("Target Classes:")
print(y.value_counts())
```

### 3.2 Split the Dataset into Training and Testing Sets

We use an 80:20 train-test split.

```python
print(y.value_counts())
print("Number of classes:", y.nunique())
print("Minimum samples in a class:", y.value_counts().min())
```

```python
print("X_tfidf shape:", X_tfidf.shape)
print("y shape:", y.shape)
```

```python
class_counts = y.value_counts()

print(class_counts[class_counts < 2])
```

```python
print("y shape:", y.shape)
print("Number of unique classes:", y.nunique())

counts = y.value_counts()

print("\nMinimum samples:", counts.min())
print("\nClasses with minimum count:")
print(counts[counts == counts.min()])

print("\nAny missing labels?")
print(y.isnull().sum())
```

```python
from sklearn.model_selection import train_test_split
```

```python
# Keep only classes with at least 2 samples
valid_classes = y.value_counts()[y.value_counts() >= 2].index

df_filtered = df[df[target_column].isin(valid_classes)]

# Recreate X and y
X_tfidf = tfidf.fit_transform(df_filtered['processed_text'])
y = df_filtered[target_column]

# Split with stratification
X_train, X_test, y_train, y_test = train_test_split(
    X_tfidf,
    y,
    test_size=0.20,
    random_state=42,
    stratify=y
)

print("Training Data Shape:", X_train.shape)
print("Testing Data Shape:", X_test.shape)
```

### 13.3 Train a Naïve Bayes Model

Multinomial Naïve Bayes is widely used for text classification because it performs well with sparse text features.

```python
from sklearn.naive_bayes import MultinomialNB

# Initialize the model
nb_model = MultinomialNB()

# Train the model
nb_model.fit(X_train, y_train)

print("Naïve Bayes model trained successfully.")
```

###  3.4 Make Predictions Using Naïve Bayes

```python
# Predict on test data
y_pred_nb = nb_model.predict(X_test)

# Display predictions
print(y_pred_nb[:10])
```

Why It Is Used for Text Classification

Fast and computationally efficient.
Handles sparse matrices effectively.
Performs well on small and medium-sized datasets.
Frequently used in spam detection and sentiment analysis.

### 13.5 Train a Support Vector Machine (SVM)

For text classification, LinearSVC is preferred because it efficiently handles high-dimensional sparse data.

```python
from sklearn.svm import LinearSVC

# Initialize the model
svm_model = LinearSVC(random_state=42)

# Train the model
svm_model.fit(X_train, y_train)

print("SVM model trained successfully.")
```

### 13.6 Make Predictions Using SVM

```python
# Predict on test data
y_pred_svm = svm_model.predict(X_test)

# Display predictions
print(y_pred_svm[:10])
```

Why It Is Used for Text Classification

Performs exceptionally well on high-dimensional datasets.
Effective with sparse TF-IDF features.
Robust against overfitting.
Often achieves higher classification accuracy than simpler models.

###4. Model Comparison

In this section, we evaluate the performance of the trained Naïve Bayes and Support Vector Machine (SVM) models using appropriate evaluation metrics. Since emotion classification is a multi-class classification problem, metrics such as Accuracy, Precision, Recall, and F1-score provide a comprehensive assessment of model performance.

### 14.1 Import Evaluation Metrics

```python
from sklearn.metrics import (
    accuracy_score,
    classification_report,
    f1_score,
    confusion_matrix
)
```

14.2 Evaluate the Naïve Bayes Model

```python
# Accuracy
nb_accuracy = accuracy_score(y_test, y_pred_nb)

# Weighted F1-score
nb_f1 = f1_score(y_test, y_pred_nb, average='weighted')

print("Naïve Bayes Performance")
print("-" * 30)
print("Accuracy :", round(nb_accuracy, 4))
print("F1-Score :", round(nb_f1, 4))

print("\nClassification Report:")
print(classification_report(y_test, y_pred_nb))
```

### 14.3 Evaluate the SVM Model

```python
# Accuracy
svm_accuracy = accuracy_score(y_test, y_pred_svm)

# Weighted F1-score
svm_f1 = f1_score(y_test, y_pred_svm, average='weighted')

print("SVM Performance")
print("-" * 30)
print("Accuracy :", round(svm_accuracy, 4))
print("F1-Score :", round(svm_f1, 4))

print("\nClassification Report:")
print(classification_report(y_test, y_pred_svm))
```

### 14.4 Compare Both Models

```python
comparison = pd.DataFrame({
    'Model': ['Naïve Bayes', 'Support Vector Machine'],
    'Accuracy': [nb_accuracy, svm_accuracy],
    'Weighted F1-Score': [nb_f1, svm_f1]
})

comparison
```

###14.5 Identify the Best Model

```python
best_model = comparison.loc[
    comparison['Weighted F1-Score'].idxmax(),
    'Model'
]

print("Best Performing Model:", best_model)
```
